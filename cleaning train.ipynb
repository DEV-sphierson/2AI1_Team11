# cleaning train.csv

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("train.csv")
df
```

```python
df.head()
```

## checking how many missing(NaN) values are there 

```python
df.isnull().sum()
```

## filling NaN using median

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['Embarked'])
```

```python
df.isnull().sum()
```

# drop whole column of cabin as more than 75% value are missing

```python
df=df.drop("Cabin", axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

# 2. checking duplicate values

```python
df.duplicated().sum()
```

## there are no duplicates in this dataset

# 3. check data types

```python
df.info()
```

## 4. convert categorical columns to numerical

```python
df["Sex"]=df["Sex"].map({"male":0,"female":1})
```

```python
df
```

```python
df.info()
```

```python
df["Embarked"]=df["Embarked"].map({"S":0,"C":1,"Q":2})
```

```python
df.info()
```

## 5. we will remove ticket column because it has too much unique values,no useful pattern,adds noise to ML model(nose=useless information in data that does not helps model learn patterns

```python
df=df.drop("Ticket",axis=1)
```

```python
df
```

```python
df.info()
```

## remove the colunm- name as it  adds noise

```python
df=df.drop("Name",axis=1)
```

```python
df.info()
```
